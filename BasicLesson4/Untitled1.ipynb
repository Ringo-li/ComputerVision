```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
black_img = np.zeros(shape=(800,800,3),dtype=np.int16)
```

```python
plt.imshow(black_img)
```

```python
cv2.rectangle(black_img,pt1=(200,200),pt2=(700,400),thickness=10,color=(0,255,0))
```

```python
plt.imshow(black_img)
```

```python
cv2.circle(black_img,center=(550,550),radius=150,color=(0,0,255),thickness=10)
```

```python
plt.imshow(black_img)
```

```python
cv2.circle(img=black_img,center=(550,550),radius=50,color=(0,0,255),thickness=-1)
```

```python
plt.imshow(black_img)
```

```python
cv2.line(img=black_img,pt1=(100,100),pt2=(700,700),color=(255,255,0),thickness=10)
```

```python
plt.imshow(black_img)
```

```python
font = cv2.FONT_HERSHEY_PLAIN
```

```python
cv2.putText(img=black_img,text="python",org=(300,150),fontFace=font,fontScale=4,color=(255,0,255),thickness=10)
```

```python
plt.imshow(black_img)
```
